# Defenisi Eigenvalue Eigenvector
---


**Eigenvector** (vektor eigen) dari suatu matriks kuadrat $A$ adalah vektor tak nol $\mathbf{v}$ sedemikian sehingga:

$$
A\mathbf{v} = \lambda \mathbf{v}
$$

di mana $\lambda$ adalah sebuah skalar yang disebut **eigenvalue** (nilai eigen) yang bersesuaian dengan vektor $\mathbf{v}$.



Intuisi

- Matriks $A$ mentransformasikan vektor $\mathbf{v}$.
- Jika hasil transformasi hanya mengubah panjang atau membalik arah dari $\mathbf{v}$ (tetap berada pada garis yang sama), maka $\mathbf{v}$ adalah **eigenvector** dan $\lambda$ adalah **eigenvalue**.



Contoh

Misalkan kita punya matriks:




$$
A = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}, \quad \mathbf{v} = \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$



Hitung:



$$
A\mathbf{v} = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix} \begin{bmatrix} 1 \\ 1 \end{bmatrix} = \begin{bmatrix} 3 \\ 3 \end{bmatrix} = 3 \cdot \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$



Maka:

- $\mathbf{v} = \begin{bmatrix} 1 \\ 1 \end{bmatrix}$ adalah **eigenvector**
- $\lambda = 3$ adalah **eigenvalue**



 Kegunaan Eigenvalue dan Eigenvector

- Analisis sistem dinamis (fisika, ekonomi)
- Reduksi dimensi (seperti PCA)
- Pengenalan wajah/gambar
- Graph theory
- Sistem persamaan diferensial





```python
import numpy as np

# Definisikan matriks A
A = np.array([[2, 1],
              [1, 2]])

# Hitung eigenvalue dan eigenvector
eigenvalues, eigenvectors = np.linalg.eig(A)

# Tampilkan hasil
print("Matriks A:")
print(A)
print("\nEigenvalue:")
print(eigenvalues)
print("\nEigenvector (kolom-kolom):")
print(eigenvectors)
```

**Apakah Eigenvector Tersebut Ortogonal dan Ortonormal?**

Diberikan dua buah eigenvector hasil dari dekomposisi eigen:




$$
v_1 = \begin{bmatrix} \frac{1}{\sqrt{2}} \\[6pt] \frac{1}{\sqrt{2}} \end{bmatrix}, \quad
v_2 = \begin{bmatrix} -\frac{1}{\sqrt{2}} \\[6pt] \frac{1}{\sqrt{2}} \end{bmatrix}
$$




---

#### 1. Apakah **ortogonal**?

Kita hitung hasil dot product:




$$
v_1 \cdot v_2 = \left( \frac{1}{\sqrt{2}} \right)\left( -\frac{1}{\sqrt{2}} \right) + \left( \frac{1}{\sqrt{2}} \right)\left( \frac{1}{\sqrt{2}} \right)
= -\frac{1}{2} + \frac{1}{2} = 0
$$




Karena $v_1 \cdot v_2 = 0$, maka:




$$
\text{v}_1 \text{ dan } \text{v}_2 \text{ saling ortogonal}
$$





---

#### 2. Apakah **ortonormal**?

Hitung panjang (norma) masing-masing vektor:





$$
\|v_1\| = \sqrt{ \left( \frac{1}{\sqrt{2}} \right)^2 + \left( \frac{1}{\sqrt{2}} \right)^2 }
= \sqrt{ \frac{1}{2} + \frac{1}{2} } = \sqrt{1} = 1
$$




$$
\|v_2\| = \sqrt{ \left( -\frac{1}{\sqrt{2}} \right)^2 + \left( \frac{1}{\sqrt{2}} \right)^2 }
= \sqrt{ \frac{1}{2} + \frac{1}{2} } = \sqrt{1} = 1
$$




Karena kedua vektor saling ortogonal dan bernorma 1, maka:




$$
\text{v}_1 \text{ dan } \text{v}_2 \text{ adalah ortonormal}
$$




---

### Kesimpulan:

Dua eigenvector dari matriks




$$
A = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}
$$




adalah:

- **Ortogonal** karena $v_1 \cdot v_2 = 0$
- **Ortonormal** karena $\|v_1\| = \|v_2\| = 1$


## Mencari Eigenvalue dan Eigenvector
-------
image.png


Diberikan matriks:




$$
A = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}
$$



Langkah 1: Cari eigenvalue ($\lambda$)

Kita gunakan persamaan karakteristik:



$$
\det(A - \lambda I) = 0
$$




Hitung:




$$
\begin{align*}
\det \left( \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix} - \lambda \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} \right)
&= \det \left( \begin{bmatrix} 2 - \lambda & 1 \\ 1 & 2 - \lambda \end{bmatrix} \right) \\
&= (2 - \lambda)^2 - 1 = 0 \\
&= \lambda^2 - 4\lambda + 3 = 0
\end{align*}
$$




Penyelesaian:




$$
\lambda = 1, \quad \lambda = 3
$$





Langkah 2: Cari eigenvector untuk setiap eigenvalue
Untuk $\lambda = 1$:

Kita selesaikan:




$$
(A - I)\mathbf{v} = 0
$$





$$
\left( \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix} - \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} \right) \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 1 & 1 \\ 1 & 1 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}
$$




Sehingga:





$$
x + y = 0 \Rightarrow y = -x
$$





Eigenvector-nya:




$$
\mathbf{v}_1 = \begin{bmatrix} 1 \\ -1 \end{bmatrix}
$$





Untuk $\lambda = 3$:




$$
(A - 3I)\mathbf{v} = 0
$$





$$
\left( \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix} - \begin{bmatrix} 3 & 0 \\ 0 & 3 \end{bmatrix} \right) \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} -1 & 1 \\ 1 & -1 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}
$$




Sehingga:




$$
- x + y = 0 \Rightarrow y = x
$$



Eigenvector-nya:



$$
\mathbf{v}_2 = \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$

### Definisi Ortogonal dan Ortonormal

Misalkan kita memiliki dua vektor $v_1$ dan $v_2$.

---

#### Ortogonal

$v_1$ dan $v_2$ dikatakan **ortogonal** jika hasil perkalian titik (dot product) mereka adalah nol:





$$
v_1 \cdot v_2 = 0
$$




---

#### Ortonormal

$v_1$ dan $v_2$ dikatakan **ortonormal** jika:

1. Mereka ortogonal:




$$
v_1 \cdot v_2 = 0
$$



2. Masing-masing vektor memiliki panjang (norma) satu:




$$
\|v_1\| = \sqrt{v_1 \cdot v_1} = 1, \quad \|v_2\| = \sqrt{v_2 \cdot v_2} = 1
$$




---

#### Contoh pada eigenvector

Misalkan:





$$
v_1 = \begin{bmatrix} 2 \\ 1 \end{bmatrix}, \quad v_2 = \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$




Hitung dot product:





$$
v_1 \cdot v_2 = 2 \times 1 + 1 \times 1 = 3 \neq 0
$$




Jadi, $v_1$ dan $v_2$ **tidak ortogonal**.

---

#### Normalisasi vektor

Untuk mendapatkan vektor ortonormal, vektor harus dinormalisasi menjadi panjang 1:






$$
\hat{v}_1 = \frac{v_1}{\|v_1\|} = \frac{1}{\sqrt{2^2 + 1^2}} \begin{bmatrix} 2 \\ 1 \end{bmatrix} = \frac{1}{\sqrt{5}} \begin{bmatrix} 2 \\ 1 \end{bmatrix}
$$





$$
\hat{v}_2 = \frac{v_2}{\|v_2\|} = \frac{1}{\sqrt{1^2 + 1^2}} \begin{bmatrix} 1 \\ 1 \end{bmatrix} = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$




---

#### Apakah $\hat{v}_1$ dan $\hat{v}_2$ ortogonal?

Hitung dot product:




$$
\hat{v}_1 \cdot \hat{v}_2 = \frac{1}{\sqrt{5}} \frac{1}{\sqrt{2}} (2 \times 1 + 1 \times 1) = \frac{3}{\sqrt{10}} \neq 0
$$




Maka, mereka **tidak ortogonal**, sehingga **tidak ortonormal**.

---

### Kesimpulan

Eigenvector matriks di atas **tidak ortogonal maupun ortonormal secara langsung**.  
Untuk mendapatkan basis ortonormal, dapat dilakukan proses Gram-Schmidt.
