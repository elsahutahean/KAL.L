## *Sistem Persamaan Linier*

Pengertian Dasar Sistem persamaan linear adalah kumpulan dari dua atau lebih persamaan yang memiliki variabel yang sama. Tujuannya adalah mencari nilai
variabel-variabel tersebut sehingga semua persamaan dalam sistem tersebut terpenuhi.
Sifat Persamaan Linear
Ada beberapa sifat-sifat yang dimiliki oleh persamaan linear, yaitu sebagai berikut:

- Penjumlahan dan pengurangan bilangan kedua ruas tak akan mengubah persamaan nilai.
- Perkalian dan pembagian bilangan kedua ruas tidak mengubah nilai persamaan
- Nilai persamaan tidak berubah jika kedua ruas ditambah atau dikurangi bilangan yang sama.
- Suatu persamaan jika dipindah ruas maka penjumlahan berubah jadi pengurangan, perkalian berubah menjadi Pembagian, dan sebaliknya.

1. Persamaan Linear Satu Variabel
Sesuai dengan namanya, persamaan linear satu variabel hanya berisi satu variabel berangka 1 yang berbentuk kalimat terbuka dengan dihubungkan tanda =.
2. Persamaan Dua Variabel Linier
Sesuai dengan namanya , Persamaan Linear Dua Variabel merupakan sistem persamaan dengan variabel yang berisi dua dengan tingkat 1. Persamaan Linear Dua Variabel menggunakan relasi = dan tidak ada perkalian variabel di setiap persamaan.
3. Persamaan Linear Tiga Variabel
Persamaan linear tiga variabel merupakan bentuk perluasan dari persamaan linear dua variabel. Sama seperti persamaan dua variabel linier, persamaan ini juga dapat diselesaikan dengan dua metode, yaitu substitusi dan eliminasi.




---

Berikut adalah penyelesaian sistem persamaan linear tiga variabel menggunakan metode eliminasi Gauss dengan format yang lebih rapi dan mudah dipahami.  

---

**1. Kasus Satu Solusi (Solusi Unik)**  

Diketahui sistem persamaan:  




$$   
\begin{aligned}
x + y + z &= 6 \\
2x + 3y + z &= 14 \\ 
y + 2z &= 8
\end{aligned}
$$




Konversi ke matriks augmented:




$$
\begin{bmatrix}
1 & 1 & 1 &  6 \\
2 & 3 & 1 &  14 \\
0 & 1 & 2 &  8
\end{bmatrix}
$$




**Langkah-langkah Eliminasi Gauss:**
1. Eliminasi elemen di bawah pivot pertama ( x ):  
   - Baris 2 → Baris 2 - 2(Baris 1)




$$
\begin{bmatrix}
1 & 1 & 1 & 6 \\
0 & 1 & -1 & 2 \\
0 & 1 & 2 & 8
\end{bmatrix}
$$




2. Eliminasi elemen di bawah pivot kedua (\( y \)):  
   - Baris 3 → Baris 3 - Baris 2 




$$
\begin{bmatrix}
1 & 1 & 1 & 6 \\
0 & 1 & -1 & 2 \\
0 & 0 & 3 & 6
\end{bmatrix}
$$




3. **Substitusi balik:**  
   - Dari baris ketiga: $( 3z = 6 \Rightarrow z = 2 )$ 
   - Dari baris kedua: $( y - 2 = 2 \Rightarrow y = 4 )$  
   - Dari baris pertama: $( x + 4 + 2 = 6 \Rightarrow x = 0 )$  



**Hasil:**  

(x, y, z) = (0, 4, 2)
 

---

**2. Kasus Tak Hingga Solusi**  

Diketahui sistem persamaan:  




$$
\begin{aligned}
x + y + z &= 6 \\
2x + 2y + 2z &= 12 \\
3x + 3y + 3z &= 18
\end{aligned}
$$ 




Konversi ke matriks augmented:  



$$
\begin{bmatrix}
1 & 1 & 1 &  6 \\
2 & 2 & 2 &  12 \\
3 & 3 & 3 &  18
\end{bmatrix}
$$ 




**Langkah-langkah Eliminasi Gauss:**
1. Eliminasi elemen di bawah pivot pertama:  
   - Baris 2 → Baris 2 - 2(Baris 1)  
   - Baris 3 → Baris 3 - 3(Baris 1)




$$
\begin{bmatrix}
1 & 1 & 1 &  6 \\
0 & 0 & 0 &  0 \\
0 & 0 & 0 &  0
\end{bmatrix}
$$




Karena ada **baris nol sepenuhnya**, sistem ini memiliki **tak hingga banyak solusi**.  

2. **Penyelesaian dalam bentuk parameter:**  
   - Misalkan $( z = t )$ (variabel bebas)  
   - Dari baris pertama: $( x + y + t = 6 \Rightarrow x + y = 6 - t )$ 
   - Misalkan $( y = s ), maka ( x = 6 - s - t )$  

**Hasil (dalam bentuk parameter):**  



$$
(x, y, z) = (6 - s - t, s, t), \quad s, t \in \mathbb{R}
$$ 



---

 **3. Kasus Tanpa Solusi**  

Diketahui sistem persamaan:





$$
\begin{aligned}
x + y + z &= 6 \\
2x + 3y + z &= 14 \\
x + y + z &= 8
\end{aligned}
$$ 




Konversi ke matriks augmented:



$$
\begin{bmatrix}
1 & 1 & 1 &  6 \\
2 & 3 & 1 &  14 \\
1 & 1 & 1 &  8
\end{bmatrix}
$$ 




**Langkah-langkah Eliminasi Gauss:**

1. Eliminasi elemen di bawah pivot pertama:  
   - Baris 3 → Baris 3 - Baris 1 




$$
\begin{bmatrix}
1 & 1 & 1 &  6 \\
2 & 3 & 1 &  14 \\
0 & 0 & 0 &  2
\end{bmatrix}
$$




Baris ketiga menyatakan **0 = 2**, yang merupakan **kontradiksi**.  

**Kesimpulan:**  
Sistem ini **tidak memiliki solusi**.

---

**Ringkasan Hasil**
| Jenis Solusi | Bentuk Matriks Setelah Eliminasi | Kesimpulan |
|-------------|---------------------------------|------------|
| **Satu Solusi** | Pivot di setiap kolom variabel, tanpa kontradiksi | ( (x, y, z) )unik |
| **Tak Hingga Solusi** | Baris nol sepenuhnya, ada variabel bebas | Solusi berbentuk parameter |

## Ringkasan Materi: Transformasi Linear dan Matriks Standar
**Math 1410 - Halaman 225–238**

## 1. Transformasi Linear

**Definisi:**  
Sebuah fungsi $T: \mathbb{R}^n \rightarrow \mathbb{R}^m$ disebut **transformasi linear** jika untuk semua vektor $\mathbf{u}, \mathbf{v} \in \mathbb{R}^n$ dan semua skalar $c \in \mathbb{R}$, berlaku:



$$
T(\mathbf{u} + \mathbf{v}) = T(\mathbf{u}) + T(\mathbf{v}) \quad \text{(additivitas)}
$$



$$
T(c\mathbf{u}) = cT(\mathbf{u}) \quad \text{(homogenitas)}
$$




Transformasi linear selalu memetakan titik nol ke titik nol:


$$
T(\mathbf{0}) = \mathbf{0}
$$



## 2. Contoh Transformasi Linear

Beberapa contoh transformasi linear di $\mathbb{R}^2$ dan $\mathbb{R}^3$:
- Rotasi
- Refleksi
- Proyeksi
- Peregangan (scaling)

**Contoh:**  
Jika $T(x, y) = (2x + y, x - y)$, maka $T$ adalah transformasi linear karena memenuhi kedua sifat di atas.

---

## 3. Transformasi yang Bukan Linear

Contoh transformasi yang **bukan** linear:
- $T(x, y) = (x + 1, y)$ (karena ada konstanta)
- $T(x, y) = (x^2, y)$ (karena tidak linier dalam $x$)

---

## 4. Matriks Standar dari Transformasi Linear

Setiap transformasi linear $T: \mathbb{R}^n \rightarrow \mathbb{R}^m$ dapat direpresentasikan dalam bentuk matriks:


$$
T(\mathbf{x}) = A\mathbf{x}
$$



Matriks $A$ disebut **matriks standar** dari transformasi $T$.

**Cara Menentukan Matriks Standar:**
1. Hitung $T(\mathbf{e}_1), T(\mathbf{e}_2), \dots, T(\mathbf{e}_n)$
2. Susun hasilnya sebagai kolom-kolom pada matriks $A$

**Contoh:**
Jika $T(x, y) = (2x + y, x - y)$, maka:
- $T(1, 0) = (2, 1)$
- $T(0, 1) = (1, -1)$

Maka:



$$
A = \begin{pmatrix}
2 & 1 \\
1 & -1
\end{pmatrix}
$$




## 5. Komposisi Transformasi Linear

Jika $T_1$ dan $T_2$ adalah transformasi linear dengan matriks $A$ dan $B$, maka komposisi $T = T_2 \circ T_1$ juga linear, dengan matriks:




$$
T(\mathbf{x}) = B(A\mathbf{x}) = (BA)\mathbf{x}
$$





## 6. Sifat Transformasi Linear

- $T(\mathbf{0}) = \mathbf{0}$
- $T$ memetakan kombinasi linear menjadi kombinasi linear:
  $$
  T(c_1\mathbf{v}_1 + c_2\mathbf{v}_2 + \dots + c_k\mathbf{v}_k) = c_1T(\mathbf{v}_1) + c_2T(\mathbf{v}_2) + \dots + c_kT(\mathbf{v}_k)
  $$

---

## 7. Geometri Transformasi Linear

Transformasi linear di $\mathbb{R}^2$ dapat:
- Merefleksikan terhadap garis
- Memutar vektor
- Memperbesar atau mengecilkan bentuk
- Memproyeksikan ke garis atau bidang

**Contoh Geometris:**
- $T(x, y) = (y, x)$ ⟹ refleksi terhadap garis $y = x$
- $T(x, y) = (-y, x)$ ⟹ rotasi 90° CCW
- $T(x, y) = (x, 0)$ ⟹ proyeksi ke sumbu-$x$

---

## 8. Kesimpulan

Transformasi linear memungkinkan kita:
- Merepresentasikan perubahan vektor sebagai perkalian matriks
- Menganalisis sistem linear dengan cara yang efisien
- Memahami struktur aljabar dari ruang vektor

Transformasi ini merupakan dasar penting dalam geometri, fisika, komputer grafis, machine learning, dan banyak bidang lainnya.

## Pengertian Transformasi Linier

Transformasi linear merupakan fungsi yang memetakan vektor dari satu ruang vektor ke ruang vektor lain dimana memenuhi sifat-sifat tertentu seperti linearitas. Sifat-sifat transformasi linear meliputi kernel yang merupakan himpunan vektor yang dipetakan ke nilai nol, serta jangkauan yang merupakan himpunan bayangan vektor hasil transformasi. Teorema-teorema penting terkait transformasi linear antara lain mengenai dimensi kernel dll.


Transformasi linier adalah suatu fungsi:




$$
T: \mathbb{R}^n \to \mathbb{R}^m
$$





yang memenuhi dua sifat utama untuk setiap vektor $( \mathbf{u}, \mathbf{v} \in \mathbb{R}^n )$ dan skalar $( c \in \mathbb{R} )$:

1. **Aditif**:




   $$
   T(\mathbf{u} + \mathbf{v}) = T(\mathbf{u}) + T(\mathbf{v})
   $$




2. **Homogen**:




   $$
   T(c\mathbf{u}) = cT(\mathbf{u})
   $$




Jika sebuah transformasi memenuhi kedua sifat tersebut, maka transformasi tersebut disebut **transformasi linier**. Transformasi ini mempertahankan operasi penjumlahan dan perkalian skalar dari ruang vektor.





## Pembuktian Transformasi Linier

Diberikan transformasi $( T: \mathbb{R}^2 \to \mathbb{R}^2 )$ didefinisikan oleh:




$$
T(v_1, v_2) = (v_1 + v_2, v_1)
$$




Transformasi $( T )$ dikatakan linier jika:

1. $( T(\mathbf{u} + \mathbf{v}) = T(\mathbf{u}) + T(\mathbf{v}) )$
2. $( T(c\mathbf{u}) = cT(\mathbf{u}) )$

Misalkan $( \mathbf{u} = (u_1, u_2) )$, $( \mathbf{v} = (v_1, v_2) )$, dan $( c \in \mathbb{R} )$.

### 1. Penjumlahan

Langkah pertama:




$$
T(\mathbf{u} + \mathbf{v}) = T((u_1 + v_1, u_2 + v_2)) = (u_1 + u_2 + v_1 + v_2, u_1 + v_1)
$$




Langkah kedua:





$$
T(\mathbf{u}) + T(\mathbf{v}) = (u_1 + u_2, u_1) + (v_1 + v_2, v_1) = (u_1 + u_2 + v_1 + v_2, u_1 + v_1)
$$




Jadi, $( T(\mathbf{u} + \mathbf{v}) = T(\mathbf{u}) + T(\mathbf{v}) )$ 

### 2. Perkalian Skalar




$$
T(c\mathbf{u}) = T((cu_1, cu_2)) = (cu_1 + cu_2, cu_1) = c(u_1 + u_2, u_1) = cT(\mathbf{u})
$$



### Kesimpulan

Karena $( T )$ memenuhi kedua sifat transformasi linier, maka $( T )$ adalah transformasi linier.

## Refleksi terhadap Garis $y = 2$

Untuk merefleksikan titik-titik terhadap garis horizontal $y = 2$, kita tidak bisa langsung menggunakan satu matriks. Oleh karena itu, refleksi dilakukan melalui tiga langkah transformasi berikut:

---

### Langkah 1: Translasi Turun sebesar 2 Satuan

Kita geser seluruh titik ke bawah sejauh 2 satuan, agar garis $y = 2$ menjadi sejajar dengan sumbu-$x$ (yakni $y = 0$):




$$
\begin{bmatrix}
x 
y
\end{bmatrix}
\longrightarrow
\begin{bmatrix}
x 
y - 2
\end{bmatrix}
$$




---

### Langkah 2: Refleksi terhadap Sumbu-$x$

Setelah posisi digeser, kita lakukan refleksi terhadap sumbu-$x$ menggunakan matriks refleksi:
  


$$
R_x =
\begin{bmatrix}
1 & 0 
0 & -1
\end{bmatrix}
$$



Sehingga:



$$
\begin{bmatrix}
x 
y - 2
\end{bmatrix}
\longrightarrow
R_x \cdot
\begin{bmatrix}
x 
y - 2
\end{bmatrix}
=
\begin{bmatrix}
x 
-(y - 2)
\end{bmatrix}
$$




---

### Langkah 3: Translasi Naik sebesar 2 Satuan

Kita geser kembali hasil refleksi tadi ke atas sejauh 2 satuan:




$$
\begin{bmatrix}
x 
-(y - 2)
\end{bmatrix}
+
\begin{bmatrix}
0 
2
\end{bmatrix}
=
\begin{bmatrix}
x 
- y + 4
\end{bmatrix}
$$





---

## Rumus Umum Refleksi terhadap $y = c$

Secara umum, jika $\mathbf{v} = \begin{bmatrix} x \\ y \end{bmatrix}$ dan kita ingin mencerminkan terhadap garis horizontal $y = c$, maka rumus refleksinya adalah:





$$
\mathbf{v}' =
R_x \cdot \left( \mathbf{v} - \begin{bmatrix} 0 \\ c \end{bmatrix} \right)
+ \begin{bmatrix} 0 \\ c \end{bmatrix}
$$




atau dituliskan langsung sebagai:




$$
\mathbf{v}' =
\begin{bmatrix}
x \\
- y + 2c
\end{bmatrix}
$$




```python
import numpy as np
import matplotlib.pyplot as plt

# Titik-titik asal (domain)
x = np.array([1, 0, -1])
y = np.array([1, 0, -1])
koordinats = np.array([x, y])

# Langkah 1: Translasi turun 2 satuan
T_down = np.array([[0, 1], [1, 0]])  # tidak dipakai secara langsung, hanya ilustrasi logika
koordinats_turun = koordinats.copy()
koordinats_turun[1, :] -= 2

# Langkah 2: Refleksi terhadap sumbu x
R_x = np.array([[1, 0], [0, -1]])
koordinats_refleksi = R_x @ koordinats_turun

# Langkah 3: Translasi naik 2 satuan
koordinats_refleksi[1, :] += 2

# Ambil hasil refleksi
x_LT2 = koordinats_refleksi[0, :]
y_LT2 = koordinats_refleksi[1, :]

# Membuat gambar dan sumbu-sumbu
fig, ax = plt.subplots()

# Plot titik-titik
ax.plot(x, y, 'ro')  # titik asli
ax.plot(x_LT2, y_LT2, 'bo')  # bayangan refleksi

# Menghubungkan titik
ax.plot(x, y, 'r', ls="--")
ax.plot(x_LT2, y_LT2, 'b')

# Gambar sumbu
ax.axvline(x=0, color="k", ls=":")
ax.axhline(y=0, color="k", ls=":")
ax.axhline(y=2, color="g", ls="-")  # garis refleksi y=2
ax.grid(True)
ax.axis([-2, 2, -2, 5])
ax.set_aspect('equal')
ax.set_title("Refleksi terhadap garis y = 2")
plt.show()
```

## Refleksi terhadap Garis $x = 2$

Untuk merefleksikan titik-titik terhadap garis vertikal $x = 2$, kita tidak bisa menggunakan satu matriks saja. Maka digunakan 3 langkah transformasi:

---

### Langkah 1: Translasi ke kiri sebesar 2 satuan

Geser semua titik ke kiri agar garis $x = 2$ menjadi $x = 0$:




$$
\begin{bmatrix}
x 
y
\end{bmatrix}
\longrightarrow
\begin{bmatrix}
x - 2 
y

$$

## Rumus Umum Refleksi terhadap $x = a$

Jika $\mathbf{v} = \begin{bmatrix} x  y \end{bmatrix}$ dan kita refleksikan terhadap garis $x = a$, maka:




$$
\mathbf{v}' =
R_y \cdot \left( \mathbf{v} - \begin{bmatrix} a  0 \end{bmatrix} \right)
+ \begin{bmatrix} a  0 \end{bmatrix}
$$




Hasil akhirnya:




$$
\mathbf{v}' =
\begin{bmatrix}
- x + 2a 
y
\end{bmatrix}
$$




```python
import numpy as np
import matplotlib.pyplot as plt

# Titik-titik asal (domain)
x = np.array([1, 0, -1])
y = np.array([1, 0, -1])
koordinats = np.array([x, y])

# Langkah 1: Translasi ke kiri sejauh 2 satuan
koordinats_kiri = koordinats.copy()
koordinats_kiri[0, :] -= 2

# Langkah 2: Refleksi terhadap sumbu y
R_y = np.array([[-1, 0], [0, 1]])
koordinats_refleksi = R_y @ koordinats_kiri

# Langkah 3: Translasi kembali ke kanan sejauh 2 satuan
koordinats_refleksi[0, :] += 2

# Ambil hasil refleksi
x_LT2 = koordinats_refleksi[0, :]
y_LT2 = koordinats_refleksi[1, :]

# Gambar
fig, ax = plt.subplots()

# Titik-titik
ax.plot(x, y, 'ro')             # Titik asli
ax.plot(x_LT2, y_LT2, 'bo')     # Titik bayangan

# Hubungkan titik-titik
ax.plot(x, y, 'r', ls="--")
ax.plot(x_LT2, y_LT2, 'b')

# Sumbu dan garis refleksi
ax.axvline(x=0, color="k", ls=":")
ax.axhline(y=0, color="k", ls=":")
ax.axvline(x=2, color="g", ls="-")  # Garis refleksi x = 2

ax.grid(True)
ax.axis([-3, 4, -2, 2])
ax.set_aspect('equal')
ax.set_title("Refleksi terhadap garis x = 2")
plt.show()
```

## Refleksi terhadap Garis $y = x$

Refleksi terhadap garis miring $y = x$ dapat dilakukan langsung menggunakan satu matriks transformasi berikut:

---

### Matriks Refleksi terhadap $y = x$




$$
R_{y = x} =
\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix}
$$





Jika $\mathbf{v} = \begin{bmatrix} x \\ y \end{bmatrix}$, maka hasil refleksi terhadap $y = x$ adalah:




$$
\mathbf{v}' =
R_{y = x} \cdot \mathbf{v}
=
\begin{bmatrix}
y \\
x
\end{bmatrix}
$$




---

### Penjelasan Geometris

Refleksi terhadap $y = x$ artinya kita menukar komponen $x$ dan $y$ dari setiap titik. Jadi:

- Titik $(x, y)$ menjadi $(y, x)$
- Titik $(1, 2)$ menjadi $(2, 1)$
- Titik $(-3, 0)$ menjadi $(0, -3)$, dst.

---

### Rumus Umum

Jika $\mathbf{v} = \begin{bmatrix} x \\ y \end{bmatrix}$, maka:





$$
\text{Refleksi terhadap } y = x \Rightarrow \mathbf{v}' = \begin{bmatrix} y \\ x \end{bmatrix}
$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Titik-titik asal (domain)
x = np.array([1, 0, -1])
y = np.array([1, 0, -1])
koordinats = np.array([x, y])

# Matriks refleksi terhadap garis y = x
R_yx = np.array([[0, 1],
                 [1, 0]])

# Transformasi (refleksi)
koordinats_refleksi = R_yx @ koordinats

# Ambil hasil bayangan
x_LT = koordinats_refleksi[0, :]
y_LT = koordinats_refleksi[1, :]

# Gambar
fig, ax = plt.subplots()

# Titik-titik
ax.plot(x, y, 'ro')             # Titik asli
ax.plot(x_LT, y_LT, 'bo')       # Titik bayangan

# Hubungkan titik-titik
ax.plot(x, y, 'r', ls="--")
ax.plot(x_LT, y_LT, 'b')

# Gambar sumbu dan garis refleksi
ax.axvline(x=0, color="k", ls=":")
ax.axhline(y=0, color="k", ls=":")
ax.plot([-2, 2], [-2, 2], 'g-', label='y = x')  # Garis refleksi

ax.grid(True)
ax.axis([-2, 2, -2, 2])
ax.set_aspect('equal')
ax.set_title("Refleksi terhadap garis y = x")
plt.legend()
plt.show()
```

# Defenisi Eigenvalue Eigenvector
---


**Eigenvector** (vektor eigen) dari suatu matriks kuadrat $A$ adalah vektor tak nol $\mathbf{v}$ sedemikian sehingga:

$$
A\mathbf{v} = \lambda \mathbf{v}
$$

di mana $\lambda$ adalah sebuah skalar yang disebut **eigenvalue** (nilai eigen) yang bersesuaian dengan vektor $\mathbf{v}$.



Intuisi

- Matriks $A$ mentransformasikan vektor $\mathbf{v}$.
- Jika hasil transformasi hanya mengubah panjang atau membalik arah dari $\mathbf{v}$ (tetap berada pada garis yang sama), maka $\mathbf{v}$ adalah **eigenvector** dan $\lambda$ adalah **eigenvalue**.



Contoh

Misalkan kita punya matriks:




$$
A = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}, \quad \mathbf{v} = \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$



Hitung:



$$
A\mathbf{v} = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix} \begin{bmatrix} 1 \\ 1 \end{bmatrix} = \begin{bmatrix} 3 \\ 3 \end{bmatrix} = 3 \cdot \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$



Maka:

- $\mathbf{v} = \begin{bmatrix} 1 \\ 1 \end{bmatrix}$ adalah **eigenvector**
- $\lambda = 3$ adalah **eigenvalue**



 Kegunaan Eigenvalue dan Eigenvector

- Analisis sistem dinamis (fisika, ekonomi)
- Reduksi dimensi (seperti PCA)
- Pengenalan wajah/gambar
- Graph theory
- Sistem persamaan diferensial





```python
import numpy as np

# Definisikan matriks A
A = np.array([[2, 1],
              [1, 2]])

# Hitung eigenvalue dan eigenvector
eigenvalues, eigenvectors = np.linalg.eig(A)

# Tampilkan hasil
print("Matriks A:")
print(A)
print("\nEigenvalue:")
print(eigenvalues)
print("\nEigenvector (kolom-kolom):")
print(eigenvectors)
```

Karena kedua vektor saling ortogonal dan bernorma 1, maka:

### Definisi Ortogonal dan Ortonormal

Misalkan kita memiliki dua vektor $v_1$ dan $v_2$.

---

#### Ortogonal

$v_1$ dan $v_2$ dikatakan **ortogonal** jika hasil perkalian titik (dot product) mereka adalah nol:





$$
v_1 \cdot v_2 = 0
$$




---

#### Ortonormal

$v_1$ dan $v_2$ dikatakan **ortonormal** jika:

1. Mereka ortogonal:




$$
v_1 \cdot v_2 = 0
$$



2. Masing-masing vektor memiliki panjang (norma) satu:




$$
\|v_1\| = \sqrt{v_1 \cdot v_1} = 1, \quad \|v_2\| = \sqrt{v_2 \cdot v_2} = 1
$$




---

#### Contoh pada eigenvector

Misalkan:





$$
v_1 = \begin{bmatrix} 2 \\ 1 \end{bmatrix}, \quad v_2 = \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$




Hitung dot product:





$$
v_1 \cdot v_2 = 2 \times 1 + 1 \times 1 = 3 \neq 0
$$




Jadi, $v_1$ dan $v_2$ **tidak ortogonal**.

---

#### Normalisasi vektor

Untuk mendapatkan vektor ortonormal, vektor harus dinormalisasi menjadi panjang 1:






$$
\hat{v}_1 = \frac{v_1}{\|v_1\|} = \frac{1}{\sqrt{2^2 + 1^2}} \begin{bmatrix} 2 \\ 1 \end{bmatrix} = \frac{1}{\sqrt{5}} \begin{bmatrix} 2 \\ 1 \end{bmatrix}
$$





$$
\hat{v}_2 = \frac{v_2}{\|v_2\|} = \frac{1}{\sqrt{1^2 + 1^2}} \begin{bmatrix} 1 \\ 1 \end{bmatrix} = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$




---

#### Apakah $\hat{v}_1$ dan $\hat{v}_2$ ortogonal?

Hitung dot product:




$$
\hat{v}_1 \cdot \hat{v}_2 = \frac{1}{\sqrt{5}} \frac{1}{\sqrt{2}} (2 \times 1 + 1 \times 1) = \frac{3}{\sqrt{10}} \neq 0
$$




Maka, mereka **tidak ortogonal**, sehingga **tidak ortonormal**.

---

### Kesimpulan

Eigenvector matriks di atas **tidak ortogonal maupun ortonormal secara langsung**.  
Untuk mendapatkan basis ortonormal, dapat dilakukan proses Gram-Schmidt.


# Perkalian Titik (Dot Product)

Dalam aljabar vektor, **perkalian titik** atau **dot product** adalah operasi aljabar antara dua vektor yang menghasilkan **skalar**. Dot product menunjukkan seberapa "sejajar" dua vektor satu sama lain.

## Rumus Umum

Jika $\vec{a}$ dan $\vec{b}$ adalah dua vektor dalam ruang tiga dimensi, maka:

### 1. Berdasarkan Komponen:

$$
\vec{a} \cdot \vec{b} = a_1b_1 + a_2b_2 + a_3b_3
$$

### 2. Berdasarkan Sudut Di Antara Vektor:

$$
\vec{a} \cdot \vec{b} = |\vec{a}| \cdot |\vec{b}| \cdot \cos \theta
$$

dengan:
- $|\vec{a}|$ dan $|\vec{b}|$ adalah panjang (norma) dari vektor $\vec{a}$ dan $\vec{b}$
- $\theta$ adalah sudut di antara kedua vektor

---

# Contoh Kasus Berdasarkan Gambar

Screenshot 2025-05-27 085521.png



Gambar menunjukkan beberapa titik dan vektor di ruang 3 dimensi (dalam konteks koordinat Cartesian).

### Diketahui:

- Titik $D = (3, 0, 0)$
- Titik $E = (-2, -1, -4)$  
  Maka vektor $\vec{a} = \vec{DE} = \begin{pmatrix} -5 \\ -1 \\ -4 \end{pmatrix}$
- Titik $A = (0, 0, 0)$  
- Titik $F = (0, 3.2, 0)$  
  Maka vektor $\vec{b} = \vec{AF} = \begin{pmatrix} 0 \\ 3.2 \\ 0 \end{pmatrix}$

---

## Langkah 1: Hitung Dot Product

Gunakan rumus komponen:

$$
\vec{a} \cdot \vec{b} = (-5)(0) + (-1)(3.2) + (-4)(0) = 0 - 3.2 + 0 = -3.2
$$

Jadi, hasil dari dot product adalah:

$$
\vec{a} \cdot \vec{b} = -3.2
$$

---

## Langkah 2: Interpretasi Geometris

- Karena $\vec{a} \cdot \vec{b} < 0$, maka sudut antara kedua vektor adalah **lebih dari 90°** (tumpul).
- Jika $\vec{a} \cdot \vec{b} = 0$, maka vektor saling tegak lurus (**orthogonal**).

> **Pada gambar**, kita juga bisa lihat bahwa sudut antara vektor $\vec{AF}$ dan $\vec{FD}$ adalah **90°**, yang berarti:
>
> $$\vec{AF} \cdot \vec{FD} = 0$$

---

## Langkah 3: Panjang (Norma) Vektor

Untuk melengkapi analisis, kita bisa hitung panjang dari kedua vektor:

### Panjang $\vec{a}$:

$$
|\vec{a}| = \sqrt{(-5)^2 + (-1)^2 + (-4)^2} = \sqrt{25 + 1 + 16} = \sqrt{42}
$$

### Panjang $\vec{b}$:

$$
|\vec{b}| = \sqrt{0^2 + 3.2^2 + 0^2} = \sqrt{10.24} = 3.2
$$

---

## Langkah 4: Gunakan Dot Product untuk Cari Sudut

Gunakan rumus:

$$
\vec{a} \cdot \vec{b} = |\vec{a}| \cdot |\vec{b}| \cdot \cos \theta
$$

Substitusikan nilai:

$$
-3.2 = \sqrt{42} \cdot 3.2 \cdot \cos \theta
$$

Bagi kedua sisi dengan $3.2$:

$$
-1 = \sqrt{42} \cdot \cos \theta
$$

Lanjut:

$$
\cos \theta = \frac{-1}{\sqrt{42}} \Rightarrow \theta = \cos^{-1} \left( \frac{-1}{\sqrt{42}} \right) \approx 98.41^\circ
$$

---

# Kesimpulan

- Dot product digunakan untuk mengetahui sudut antara dua vektor.
- Jika hasil dot product positif, sudutnya lancip (< 90°).
- Jika hasilnya negatif, sudutnya tumpul (> 90°).
- Jika hasilnya nol, vektor saling tegak lurus (90°).
- Contoh di gambar menunjukkan dot product negatif, sehingga sudut antara vektor $\vec{DE}$ dan $\vec{AF}$ adalah **tumpul**.



# Defenisi Kali Silang (Cross Product)

Hasil kali silang (cross product) adalah operasi pada dua vektor di ruang tiga dimensi yang menghasilkan vektor baru yang tegak lurus terhadap kedua vektor tersebut.

---

## Menggunakan Determinan untuk Menemukan Perkalian Silang

Pertama, kita membentuk array $3 \times 3$ seperti yang ditunjukkan di bawah ini:



$$[
\vec{u} \times \vec{v} =
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
u_1 & u_2 & u_3 \\
v_1 & v_2 & v_3
\end{vmatrix}
]$$



Kemudian kita hitung determinan tersebut menggunakan ekspansi kofaktor:




$$[
\vec{u} \times \vec{v} =
(u_2v_3 - u_3v_2)\hat{i} -
(u_1v_3 - u_3v_1)\hat{j} +
(u_1v_2 - u_2v_1)\hat{k}
]$$



Atau dapat ditulis sebagai vektor:



$$[
\vec{u} \times \vec{v} =
\begin{bmatrix}
u_2v_3 - u_3v_2 \\
-(u_1v_3 - u_3v_1) \\
u_1v_2 - u_2v_1
\end{bmatrix}
]$$


---

### Perkalian Silang $\vec{v} \times \vec{u}$


$$[
\vec{v} \times \vec{u} =
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
v_1 & v_2 & v_3 \\
u_1 & u_2 & u_3
\end{vmatrix}
]$$



$$[
= \begin{vmatrix}
v_2 & v_3 \\
u_2 & u_3
\end{vmatrix} \hat{i}
- \begin{vmatrix}
v_1 & v_3 \\
u_1 & u_3
\end{vmatrix} \hat{j}
+ \begin{vmatrix}
v_1 & v_2 \\
u_1 & u_2
\end{vmatrix} \hat{k}
]$$




$$[
= (v_2u_3 - v_3u_2)\hat{i}
- (v_1u_3 - v_3u_1)\hat{j}
+ (v_1u_2 - v_2u_1)\hat{k}
]$$




Atau sebagai vektor:



$$[
\vec{v} \times \vec{u} =
\begin{bmatrix}
v_2u_3 - v_3u_2 \\
-(v_1u_3 - v_3u_1) \\
v_1u_2 - v_2u_1
\end{bmatrix}
]$$


---

## Verifikasi dengan Vektor Satuan

Misal:



$$[
\vec{u} = (1, 0, 0), \quad \vec{v} = (0, 1, 0)
]$$



$$[
\vec{v} \times \vec{u} =
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
0 & 1 & 0 \\
1 & 0 & 0
\end{vmatrix}
]$$



$$[
= \hat{i}(1 \cdot 0 - 0 \cdot 0)
- \hat{j}(0 \cdot 0 - 0 \cdot 1)
+ \hat{k}(0 \cdot 0 - 1 \cdot 1)
= 0\hat{i} - 0\hat{j} - 1\hat{k} = -\hat{k}
]$$



---

## Rumus Perkalian Silang

- $\hat{i} \times \hat{j} = \hat{k}$
- $\hat{j} \times \hat{i} = -\hat{k}$
- $\hat{i} \times \hat{j} = -(\hat{j} \times \hat{i})$
- $\hat{i} \times \hat{k} = -\hat{j}$
- $\hat{k} \times \hat{i} = \hat{j}$
- $\hat{i} \times \hat{j} \neq \hat{j} \times \hat{i}$

---
## Latihan:

Dengan nilai:

* $\vec{v} = (v_1, v_2, v_3) = (1, \mathbf{2}, 1)$
* $\vec{u} = (u_1, u_2, u_3) = (1, 1, 1)$

Kita hitung:




$$
\vec{u} \times \vec{v} =
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
1 & 1 & 1 \\
1 & 2 & 1
\end{vmatrix}
$$




Ekspansi determinan:





$$
= \left| \begin{array}{cc}
1 & 1 \\
2 & 1
\end{array} \right| \hat{i}
-
\left| \begin{array}{cc}
1 & 1 \\
1 & 1
\end{array} \right| \hat{j}
+
\left| \begin{array}{cc}
1 & 1 \\
1 & 2
\end{array} \right| \hat{k}
$$




Hitung masing-masing:

* $\left| \begin{array}{cc} 1 & 1 \\ 2 & 1 \end{array} \right| = (1)(1) - (2)(1) = 1 - 2 = -1$
* $\left| \begin{array}{cc} 1 & 1 \\ 1 & 1 \end{array} \right| = (1)(1) - (1)(1) = 0$
* $\left| \begin{array}{cc} 1 & 1 \\ 1 & 2 \end{array} \right| = (1)(2) - (1)(1) = 2 - 1 = 1$

---

### Hasil Akhir:




$$
\vec{u} \times \vec{v} = (-1)\hat{i} - (0)\hat{j} + (1)\hat{k} = -\hat{i} + \hat{k}
$$





Atau dalam bentuk vektor:




$$
\vec{u} \times \vec{v} =
\begin{bmatrix}
-1 \\
0 \\
1
\end{bmatrix}
$$

## *Ringkasan Materi: Perkalian Silang (Cross Product)*



Perkalian silang adalah operasi antara dua vektor dalam ruang 3 dimensi (3D) yang menghasilkan *vektor baru* yang *tegak lurus (ortogonal)* terhadap kedua vektor awal.



### *Rumus Perkalian Silang*

Jika dua vektor:

* *A* = ⟨a₁, a₂, a₃⟩
* *B* = ⟨b₁, b₂, b₃⟩

Maka:





$$
\mathbf{A} \times \mathbf{B} =
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
a_1 & a_2 & a_3 \\
b_1 & b_2 & b_3 \\
\end{vmatrix} =
\mathbf{i}(a_2b_3 - a_3b_2) - 
\mathbf{j}(a_1b_3 - a_3b_1) + 
\mathbf{k}(a_1b_2 - a_2b_1)
$$




Hasilnya adalah vektor baru:



$$
\mathbf{A} \times \mathbf{B} = ⟨a_2b_3 - a_3b_2,\; -(a_1b_3 - a_3b_1),\; a_1b_2 - a_2b_1⟩
$$





### *Sifat-Sifat Perkalian Silang*

1. *Antikomutatif*





   $$
   \mathbf{A} \times \mathbf{B} = -(\mathbf{B} \times \mathbf{A})
   $$





2. *Distribusi terhadap Penjumlahan Vektor*




   $$
   \mathbf{A} \times (\mathbf{B} + \mathbf{C}) = \mathbf{A} \times \mathbf{B} + \mathbf{A} \times \mathbf{C}
   $$




3. *Perkalian silang dengan diri sendiri = nol*

   
   
   
   
   $$
   \mathbf{A} \times \mathbf{A} = \mathbf{0}
   $$






4. *Hasilnya selalu tegak lurus terhadap A dan B*




### *Implementasi dalam Kehidupan Nyata*

Perkalian silang digunakan untuk:

* Menentukan arah normal bidang (tegak lurus)
* Menghitung momen gaya dalam fisika
* Menentukan luas jajar genjang atau volume bangun ruang 3D
* Grafik dan animasi 3D (dalam komputer grafis)





### *Volume Paralelepiped (Balok Miring)*

Volume dari bangun ruang yang dibentuk oleh tiga vektor *A, **B, dan **C*:



$$
V = |\mathbf{A} \cdot (\mathbf{B} \times \mathbf{C})|
$$





##  *Soal 1*

*Tentukan luas jajaran genjang yang ditentukan oleh:*




$$
\vec{u} = \begin{pmatrix} 1 \\ 2 \end{pmatrix}, \quad 
\vec{v} = \begin{pmatrix} 2 \\ 1 \end{pmatrix}
$$






###  Konsep:

Untuk vektor 2 dimensi, luas jajargenjang bisa dihitung dengan:




$$
\text{Luas} = |u_1 v_2 - u_2 v_1|
$$





###  Perhitungan:




$$
= |(1)(1) - (2)(2)| = |1 - 4| = |-3| = \boxed{3}
$$

##  *Soal 3*

*Tentukan luas segitiga dengan titik-titik sudut:*
$A = (0, 0, 0), \quad B = (1, 3, -1), \quad C = (2, 1, 1)$

###  Konsep:

Luas segitiga dari tiga titik di ruang 3D:

1. Buat dua vektor posisi:




   $$
   \vec{AB} = B - A = (1, 3, -1), \quad \vec{AC} = C - A = (2, 1, 1)
   $$





2. Luas = ½ × panjang dari $\vec{AB} \times \vec{AC}$

---

### Langkah:





$$
\vec{AB} \times \vec{AC} =
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
1 & 3 & -1 \\
2 & 1 & 1 \\
\end{vmatrix}
= \mathbf{i}(3×1 - (-1)×1) - \mathbf{j}(1×1 - (-1)×2) + \mathbf{k}(1×1 - 3×2)
$$








$$
= \mathbf{i}(3 + 1) - \mathbf{j}(1 + 2) + \mathbf{k}(1 - 6) = \mathbf{i}(4) - \mathbf{j}(3) + \mathbf{k}(-5)
$$






$$
\vec{AB} \times \vec{AC} = \begin{pmatrix} 4 \\ -3 \\ -5 \end{pmatrix}
$$






### Hitung panjangnya:





$$
\left\| \vec{AB} \times \vec{AC} \right\| = \sqrt{4^2 + (-3)^2 + (-5)^2} = \sqrt{16 + 9 + 25} = \sqrt{50}
$$




### Jawaban:






$$
\text{Luas segitiga} = \frac{1}{2} \times \sqrt{50} = \boxed{\frac{\sqrt{50}}{2} \approx 3{,}54}
$$




---

## *Soal 4*

*Tentukan luas segitiga dengan titik-titik sudut:*
$A = (5, 2, -1), \quad B = (3, 6, 2), \quad C = (1, 0, 4)$

---

### Langkah:




$$
\vec{AB} = B - A = (3 - 5, 6 - 2, 2 - (-1)) = (-2, 4, 3) \\
\vec{AC} = C - A = (1 - 5, 0 - 2, 4 - (-1)) = (-4, -2, 5)
$$







$$
\vec{AB} \times \vec{AC} =
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
-2 & 4 & 3 \\
-4 & -2 & 5 \\
\end{vmatrix}
= \mathbf{i}(4×5 - 3×(-2)) - \mathbf{j}(-2×5 - 3×(-4)) + \mathbf{k}(-2×(-2) - 4×(-4))
$$




$$
= \mathbf{i}(20 + 6) - \mathbf{j}(-10 + 12) + \mathbf{k}(4 + 16) = \mathbf{i}(26) - \mathbf{j}(2) + \mathbf{k}(20)
$$




$$
\vec{AB} \times \vec{AC} = \begin{pmatrix} 26 \\ -2 \\ 20 \end{pmatrix}
$$





---

### Panjang:







$$
\left\| \vec{AB} \times \vec{AC} \right\| = \sqrt{26^2 + (-2)^2 + 20^2} = \sqrt{676 + 4 + 400} = \sqrt{1080}
$$






$$
\text{Luas segitiga} = \frac{1}{2} \times \sqrt{1080} = \boxed{\frac{\sqrt{1080}}{2} \approx 16{,}43}
$$
