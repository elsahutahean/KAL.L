## *Sistem Persamaan Linier*

Pengertian Dasar Sistem persamaan linear adalah kumpulan dari dua atau lebih persamaan yang memiliki variabel yang sama. Tujuannya adalah mencari nilai
variabel-variabel tersebut sehingga semua persamaan dalam sistem tersebut terpenuhi.
Sifat Persamaan Linear
Ada beberapa sifat-sifat yang dimiliki oleh persamaan linear, yaitu sebagai berikut:

- Penjumlahan dan pengurangan bilangan kedua ruas tak akan mengubah persamaan nilai.
- Perkalian dan pembagian bilangan kedua ruas tidak mengubah nilai persamaan
- Nilai persamaan tidak berubah jika kedua ruas ditambah atau dikurangi bilangan yang sama.
- Suatu persamaan jika dipindah ruas maka penjumlahan berubah jadi pengurangan, perkalian berubah menjadi Pembagian, dan sebaliknya.

1. Persamaan Linear Satu Variabel
Sesuai dengan namanya, persamaan linear satu variabel hanya berisi satu variabel berangka 1 yang berbentuk kalimat terbuka dengan dihubungkan tanda =.
2. Persamaan Dua Variabel Linier
Sesuai dengan namanya , Persamaan Linear Dua Variabel merupakan sistem persamaan dengan variabel yang berisi dua dengan tingkat 1. Persamaan Linear Dua Variabel menggunakan relasi = dan tidak ada perkalian variabel di setiap persamaan.
3. Persamaan Linear Tiga Variabel
Persamaan linear tiga variabel merupakan bentuk perluasan dari persamaan linear dua variabel. Sama seperti persamaan dua variabel linier, persamaan ini juga dapat diselesaikan dengan dua metode, yaitu substitusi dan eliminasi.




---

Berikut adalah penyelesaian sistem persamaan linear tiga variabel menggunakan metode eliminasi Gauss dengan format yang lebih rapi dan mudah dipahami.  

---

**1. Kasus Satu Solusi (Solusi Unik)**  

Diketahui sistem persamaan:  




$$   
\begin{aligned}
x + y + z &= 6 \\
2x + 3y + z &= 14 \\ 
y + 2z &= 8
\end{aligned}
$$




Konversi ke matriks augmented:




$$
\begin{bmatrix}
1 & 1 & 1 &  6 \\
2 & 3 & 1 &  14 \\
0 & 1 & 2 &  8
\end{bmatrix}
$$




**Langkah-langkah Eliminasi Gauss:**
1. Eliminasi elemen di bawah pivot pertama ( x ):  
   - Baris 2 → Baris 2 - 2(Baris 1)




$$
\begin{bmatrix}
1 & 1 & 1 & 6 \\
0 & 1 & -1 & 2 \\
0 & 1 & 2 & 8
\end{bmatrix}
$$




2. Eliminasi elemen di bawah pivot kedua (\( y \)):  
   - Baris 3 → Baris 3 - Baris 2 




$$
\begin{bmatrix}
1 & 1 & 1 & 6 \\
0 & 1 & -1 & 2 \\
0 & 0 & 3 & 6
\end{bmatrix}
$$




3. **Substitusi balik:**  
   - Dari baris ketiga: $( 3z = 6 \Rightarrow z = 2 )$ 
   - Dari baris kedua: $( y - 2 = 2 \Rightarrow y = 4 )$  
   - Dari baris pertama: $( x + 4 + 2 = 6 \Rightarrow x = 0 )$  



**Hasil:**  

(x, y, z) = (0, 4, 2)
 

---

**2. Kasus Tak Hingga Solusi**  

Diketahui sistem persamaan:  




$$
\begin{aligned}
x + y + z &= 6 \\
2x + 2y + 2z &= 12 \\
3x + 3y + 3z &= 18
\end{aligned}
$$ 




Konversi ke matriks augmented:  



$$
\begin{bmatrix}
1 & 1 & 1 &  6 \\
2 & 2 & 2 &  12 \\
3 & 3 & 3 &  18
\end{bmatrix}
$$ 




**Langkah-langkah Eliminasi Gauss:**
1. Eliminasi elemen di bawah pivot pertama:  
   - Baris 2 → Baris 2 - 2(Baris 1)  
   - Baris 3 → Baris 3 - 3(Baris 1)




$$
\begin{bmatrix}
1 & 1 & 1 &  6 \\
0 & 0 & 0 &  0 \\
0 & 0 & 0 &  0
\end{bmatrix}
$$




Karena ada **baris nol sepenuhnya**, sistem ini memiliki **tak hingga banyak solusi**.  

2. **Penyelesaian dalam bentuk parameter:**  
   - Misalkan $( z = t )$ (variabel bebas)  
   - Dari baris pertama: $( x + y + t = 6 \Rightarrow x + y = 6 - t )$ 
   - Misalkan $( y = s ), maka ( x = 6 - s - t )$  

**Hasil (dalam bentuk parameter):**  



$$
(x, y, z) = (6 - s - t, s, t), \quad s, t \in \mathbb{R}
$$ 



---

 **3. Kasus Tanpa Solusi**  

Diketahui sistem persamaan:





$$
\begin{aligned}
x + y + z &= 6 \\
2x + 3y + z &= 14 \\
x + y + z &= 8
\end{aligned}
$$ 




Konversi ke matriks augmented:



$$
\begin{bmatrix}
1 & 1 & 1 &  6 \\
2 & 3 & 1 &  14 \\
1 & 1 & 1 &  8
\end{bmatrix}
$$ 




**Langkah-langkah Eliminasi Gauss:**

1. Eliminasi elemen di bawah pivot pertama:  
   - Baris 3 → Baris 3 - Baris 1 




$$
\begin{bmatrix}
1 & 1 & 1 &  6 \\
2 & 3 & 1 &  14 \\
0 & 0 & 0 &  2
\end{bmatrix}
$$




Baris ketiga menyatakan **0 = 2**, yang merupakan **kontradiksi**.  

**Kesimpulan:**  
Sistem ini **tidak memiliki solusi**.

---

**Ringkasan Hasil**
| Jenis Solusi | Bentuk Matriks Setelah Eliminasi | Kesimpulan |
|-------------|---------------------------------|------------|
| **Satu Solusi** | Pivot di setiap kolom variabel, tanpa kontradiksi | ( (x, y, z) )unik |
| **Tak Hingga Solusi** | Baris nol sepenuhnya, ada variabel bebas | Solusi berbentuk parameter |
| **Tanpa Solusi** | Baris dengan ( 0 = c ) (c tidak sama dengan 0) | Tidak ada solusi |

## *Penyelesaian Dengan Metode Invers Matriks*

Diberikan sistem persamaan linear:



$$
\begin{aligned}
x_1 + 2x_2 + 3x_3 &= 1 \\
4x_1 - x_2 + 2x_3 &= 4 \\
5x_1 + 3x_2 - x_3 &= 5
\end{aligned}
$$




Kita tuliskan dalam bentuk matriks:

$$
A \mathbf{x} = \mathbf{b}
$$



dengan:



$$
A =
\begin{bmatrix}
1 & 2 & 3 \\
4 & -1 & 2 \\
5 & 3 & -1
\end{bmatrix}, \quad
\mathbf{x} =
\begin{bmatrix}
x_1 \\ x_2 \\ x_3
\end{bmatrix}, \quad
\mathbf{b} =
\begin{bmatrix}
1 \\ 4 \\ 5
\end{bmatrix}
$$



Karena $( A )$ memiliki invers, kita dapat mencari solusi dengan:



$$
\mathbf{x} = A^{-1} \mathbf{b}
$$




### *1. Mencari Invers Matriks $( A )$*

Kita hitung invers dari $( A )$:


$$
A^{-1} =
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
$$




Hasil perhitungan menunjukkan bahwa solusi sistem persamaan adalah:



$$
x_1 = 1, \quad x_2 \approx -8.33 \times 10^{-17}, \quad x_3 \approx -2.78 \times 10^{-17}
$$




Nilai $( x_2 )$ dan $( x_3 )$ sangat kecil (mendekati nol), yang kemungkinan besar disebabkan oleh kesalahan pembulatan dalam perhitungan numerik. Jadi, solusi akhirnya adalah:



$$
(x_1, x_2, x_3) = (1, 0, 0)
$$



### *Kesimpulan*
Sistem persamaan ini memiliki solusi unik:



$$
\mathbf{x} =
\begin{bmatrix}
1 \\ 0 \\ 0
\end{bmatrix}
$$


## Ringkasan Materi: Transformasi Linear dan Matriks Standar
**Math 1410 - Halaman 225–238**

## 1. Transformasi Linear

**Definisi:**  
Sebuah fungsi $T: \mathbb{R}^n \rightarrow \mathbb{R}^m$ disebut **transformasi linear** jika untuk semua vektor $\mathbf{u}, \mathbf{v} \in \mathbb{R}^n$ dan semua skalar $c \in \mathbb{R}$, berlaku:



$$
T(\mathbf{u} + \mathbf{v}) = T(\mathbf{u}) + T(\mathbf{v}) \quad \text{(additivitas)}
$$



$$
T(c\mathbf{u}) = cT(\mathbf{u}) \quad \text{(homogenitas)}
$$




Transformasi linear selalu memetakan titik nol ke titik nol:


$$
T(\mathbf{0}) = \mathbf{0}
$$



## 2. Contoh Transformasi Linear

Beberapa contoh transformasi linear di $\mathbb{R}^2$ dan $\mathbb{R}^3$:
- Rotasi
- Refleksi
- Proyeksi
- Peregangan (scaling)

**Contoh:**  
Jika $T(x, y) = (2x + y, x - y)$, maka $T$ adalah transformasi linear karena memenuhi kedua sifat di atas.

---

## 3. Transformasi yang Bukan Linear

Contoh transformasi yang **bukan** linear:
- $T(x, y) = (x + 1, y)$ (karena ada konstanta)
- $T(x, y) = (x^2, y)$ (karena tidak linier dalam $x$)

---

## 4. Matriks Standar dari Transformasi Linear

Setiap transformasi linear $T: \mathbb{R}^n \rightarrow \mathbb{R}^m$ dapat direpresentasikan dalam bentuk matriks:


$$
T(\mathbf{x}) = A\mathbf{x}
$$



Matriks $A$ disebut **matriks standar** dari transformasi $T$.

**Cara Menentukan Matriks Standar:**
1. Hitung $T(\mathbf{e}_1), T(\mathbf{e}_2), \dots, T(\mathbf{e}_n)$
2. Susun hasilnya sebagai kolom-kolom pada matriks $A$

**Contoh:**
Jika $T(x, y) = (2x + y, x - y)$, maka:
- $T(1, 0) = (2, 1)$
- $T(0, 1) = (1, -1)$

Maka:



$$
A = \begin{pmatrix}
2 & 1 \\
1 & -1
\end{pmatrix}
$$




## 5. Komposisi Transformasi Linear

Jika $T_1$ dan $T_2$ adalah transformasi linear dengan matriks $A$ dan $B$, maka komposisi $T = T_2 \circ T_1$ juga linear, dengan matriks:




$$
T(\mathbf{x}) = B(A\mathbf{x}) = (BA)\mathbf{x}
$$





## 6. Sifat Transformasi Linear

- $T(\mathbf{0}) = \mathbf{0}$
- $T$ memetakan kombinasi linear menjadi kombinasi linear:
  $$
  T(c_1\mathbf{v}_1 + c_2\mathbf{v}_2 + \dots + c_k\mathbf{v}_k) = c_1T(\mathbf{v}_1) + c_2T(\mathbf{v}_2) + \dots + c_kT(\mathbf{v}_k)
  $$

---

## 7. Geometri Transformasi Linear

Transformasi linear di $\mathbb{R}^2$ dapat:
- Merefleksikan terhadap garis
- Memutar vektor
- Memperbesar atau mengecilkan bentuk
- Memproyeksikan ke garis atau bidang

**Contoh Geometris:**
- $T(x, y) = (y, x)$ ⟹ refleksi terhadap garis $y = x$
- $T(x, y) = (-y, x)$ ⟹ rotasi 90° CCW
- $T(x, y) = (x, 0)$ ⟹ proyeksi ke sumbu-$x$

---

## 8. Kesimpulan

Transformasi linear memungkinkan kita:
- Merepresentasikan perubahan vektor sebagai perkalian matriks
- Menganalisis sistem linear dengan cara yang efisien
- Memahami struktur aljabar dari ruang vektor

Transformasi ini merupakan dasar penting dalam geometri, fisika, komputer grafis, machine learning, dan banyak bidang lainnya.

## Pengertian Transformasi Linier

Transformasi linear merupakan fungsi yang memetakan vektor dari satu ruang vektor ke ruang vektor lain dimana memenuhi sifat-sifat tertentu seperti linearitas. Sifat-sifat transformasi linear meliputi kernel yang merupakan himpunan vektor yang dipetakan ke nilai nol, serta jangkauan yang merupakan himpunan bayangan vektor hasil transformasi. Teorema-teorema penting terkait transformasi linear antara lain mengenai dimensi kernel dll.


Transformasi linier adalah suatu fungsi:




$$
T: \mathbb{R}^n \to \mathbb{R}^m
$$





yang memenuhi dua sifat utama untuk setiap vektor $( \mathbf{u}, \mathbf{v} \in \mathbb{R}^n )$ dan skalar $( c \in \mathbb{R} )$:

1. **Aditif**:




   $$
   T(\mathbf{u} + \mathbf{v}) = T(\mathbf{u}) + T(\mathbf{v})
   $$




2. **Homogen**:




   $$
   T(c\mathbf{u}) = cT(\mathbf{u})
   $$




Jika sebuah transformasi memenuhi kedua sifat tersebut, maka transformasi tersebut disebut **transformasi linier**. Transformasi ini mempertahankan operasi penjumlahan dan perkalian skalar dari ruang vektor.





## Pembuktian Transformasi Linier

Diberikan transformasi $( T: \mathbb{R}^2 \to \mathbb{R}^2 )$ didefinisikan oleh:




$$
T(v_1, v_2) = (v_1 + v_2, v_1)
$$




Transformasi $( T )$ dikatakan linier jika:

1. $( T(\mathbf{u} + \mathbf{v}) = T(\mathbf{u}) + T(\mathbf{v}) )$
2. $( T(c\mathbf{u}) = cT(\mathbf{u}) )$

Misalkan $( \mathbf{u} = (u_1, u_2) )$, $( \mathbf{v} = (v_1, v_2) )$, dan $( c \in \mathbb{R} )$.

### 1. Penjumlahan

Langkah pertama:




$$
T(\mathbf{u} + \mathbf{v}) = T((u_1 + v_1, u_2 + v_2)) = (u_1 + u_2 + v_1 + v_2, u_1 + v_1)
$$




Langkah kedua:





$$
T(\mathbf{u}) + T(\mathbf{v}) = (u_1 + u_2, u_1) + (v_1 + v_2, v_1) = (u_1 + u_2 + v_1 + v_2, u_1 + v_1)
$$




Jadi, $( T(\mathbf{u} + \mathbf{v}) = T(\mathbf{u}) + T(\mathbf{v}) )$ 

### 2. Perkalian Skalar




$$
T(c\mathbf{u}) = T((cu_1, cu_2)) = (cu_1 + cu_2, cu_1) = c(u_1 + u_2, u_1) = cT(\mathbf{u})
$$



### Kesimpulan

Karena $( T )$ memenuhi kedua sifat transformasi linier, maka $( T )$ adalah transformasi linier.





## Contoh Refleksi Transformasi Matriks

Berikut adalah contoh 2 titik untuk masing-masing jenis refleksi transformasi:

### 1. Reflection about the x-axis

Matriks refleksi:




$$
\begin{bmatrix}
1 & 0 \\
0 & -1
\end{bmatrix}
$$





Contoh:




$$
(2, 3) \rightarrow (2, -3) \\
(-1, 4) \rightarrow (-1, -4)
$$

---

### 2. Reflection about the y-axis

Matriks refleksi:




$$
\begin{bmatrix}
-1 & 0 \\
0 & 1
\end{bmatrix}
$$




Contoh:




$$
(2, 3) \rightarrow (-2, 3) \\
(-4, 1) \rightarrow (4, 1)
$$




---

### 3. Reflection about the line \( y = x \)

Matriks refleksi:






$$
\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix}
$$




Contoh:



$$
(2, 5) \rightarrow (5, 2) \\
(0, -3) \rightarrow (-3, 0)
$$



---

### 4. Reflection about the line \( y = -x \)

Matriks refleksi:




$$
\begin{bmatrix}
0 & -1 \\
-1 & 0
\end{bmatrix}
$$



Contoh:




$$
(3, 1) \rightarrow (-1, -3) \\
(-2, 4) \rightarrow (-4, 2)
$$




---

### 5. Reflection about the origin

Matriks refleksi:




$$
\begin{bmatrix}
-1 & 0 \\
0 & -1
\end{bmatrix}
$$



Contoh:



$$
(3, -2) \rightarrow (-3, 2) \\
(-1, -1) \rightarrow (1, 1)
$$
